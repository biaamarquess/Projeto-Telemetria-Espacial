**Regras de decisão com portas lógicas**

Para assegurar um pouso seguro, o MGPEB adota um conjunto de regras baseadas em lógica booleana. Por meio dessas regras, o sistema analisa automaticamente as condições do módulo e do ambiente, decidindo de forma objetiva se a operação pode ser realizada ou se deve ser interrompida temporariamente.
<br><br>

**A autorização de pouso ira depender das seguintes condições:**

* Combustível ≥ 80%

* Integridade estrutural = OK

* Área de pouso disponível

* Condições atmosféricas seguras
<br>

**Expressão booleana:**

Pouso = Combustivel AND integridade AND AreaDisponivel AND CondicoesClimaticas
<br>
<br>

**AND:** todas as condições precisam ser verdadeiras

**NOT:** usado para bloquear falhas

**OR:** pode ser usado para exceções (ex: prioridade crítica)
<br>
<br>

**Exemplo lógico em Python:**

```python
if combustivel >= 80 and integridade == 1 and area_livre and clima_ok:
    print("POUSO AUTORIZADO")
else:
    print("POUSO NEGADO")
```

**Protótipo em Python (MGPEB):**

```python
# FILA DE POUSO
fila = []

def adicionar_modulo(modulo):
    fila.append(modulo)

def ordenar_por_prioridade():
    fila.sort(key=lambda x: x["prioridade"], reverse=True)

def buscar_modulo(tipo):
    for m in fila:
        if m["tipo"] == tipo:
            return m
    return None

def autorizar_pouso(modulo):
    if (modulo["combustivel"] >= 80 and
        modulo["integridade"] == 1 and
        modulo["area_livre"] == True):
        return True
    return False

# EXEMPLO
modulo = {
    "tipo": "Habitação",
    "prioridade": 4,
    "combustivel": 85,
    "integridade": 1,
    "area_livre": True
}

adicionar_modulo(modulo)
ordenar_por_prioridade()

if autorizar_pouso(modulo):
    print("Pouso autorizado")
else:
    print("Pouso negado")
```

**Modelagem matemática aplicada**

Para auxiliar na tomada de decisões durante o pouso, o MGPEB utiliza modelos matemáticos que representam o comportamento físico do módulo. Um exemplo relevante é a variação da altura ao longo do tempo durante a descida.
<br><br>

**Podemos modelar com uma função quadrática:**
<br><br>

*h(t) = -5t² + 50t +100*
<br><br>

**Onde:**

* **h(t)**= altura

* **t** = tempo

* **-5** = aceleração gravitacional (simplificada)

* **50** = velocidade inicial

* **100** = altura inicial
<br>

**Análise:**

* A altura diminui com o tempo

* Existe um ponto máximo (antes da descida real começar)

* A curva é uma parábola
<br>

**Essa função com a aplicação do MGPEB ajuda a:**

* Definir o momento ideal para ativar retrofoguetes

* Controlar a velocidade de descida

* Evitar impacto brusco no solo

**Princípios ESG na base Aurora**

A construção e operação da base Aurora Siger devem seguir princípios de responsabilidade ambiental, social e de governança (ESG). Esses princípios garantem que o uso da tecnologia seja feito de forma ética, sustentável e segura, mesmo em um ambiente extremo como Marte.
<br><br>

**Ambiental (E):**
O aspecto ambiental está relacionado à preservação e ao uso consciente dos recursos naturais, mesmo em Marte. O objetivo é minimizar impactos e garantir sustentabilidade a longo prazo.
<br><br>

* Escolha da área de pouso sem comprometer o ambiente marciano

* Uso eficiente de energia solar

* Reutilização de recursos
<br>

**Social (S):**
O aspecto social envolve o bem-estar, a segurança e a qualidade de vida dos tripulantes da base, garantindo condições justas e seguras para todos.
<br><br>

* Priorizar módulos de sobrevivência (energia e saúde)

* Garantir segurança dos tripulantes

* Distribuição justa de recursos
<br>

**Governança (G):**
A governança está relacionada à forma como as decisões são tomadas e gerenciadas, priorizando transparência, organização e responsabilidade.
<br><br>

* Decisões baseadas em dados (telemetria + MGPEB)

* Transparência nas operações

* Definição de protocolos claros de pouso
